```python
# Data Transformations
data.columns = data.columns.str.replace("_"," ")
data
```

```python
data.columns = data.columns.str.replace(" ","_")
data
```

```python
data["Age"] = data["Age"].astype(int)
data
```

```python
data["Percentage"] = data["Percentage"].astype(float)
data
```

```python
data["Total Score"] = (data["Assignment_Score"] + data["Internal_Marks"] + data["Final_Exam_Marks"])
data
```

```python
data["Total Percentage"] = data["Total_Marks"]/100*100
data
```

```python
def grade(Total_Score):
  if Total_Score >= 90:
    return "A"
  elif Total_Score >= 80:
    return "B"
  elif Total_Score >= 70:
    return "C"
  elif Total_Score >= 60:
    return "D"
  else:
    return "F"
df["Grade"] = df["Percentage"].apply(grade)
df
```

```python
# Department wise Percentage
import matplotlib.pyplot as plt

department = data.groupby("Department")["Percentage"].mean()
Colors = ["#FF7873","#FF80C7","#FF9E20","#95CCDD","#9D6638"]

plt.figure(figsize=(8,5))
plt.bar(department.index, department.values, color=Colors)
plt.xlabel("Department")
plt.ylabel("Average Percentage")
plt.title("Department wise Average Percentage",fontsize=18)
plt.grid()
plt.show()
```
